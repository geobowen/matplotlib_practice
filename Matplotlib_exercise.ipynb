## Making publication-quality figures in python (Part II): Line plot, Legends, Colors
https://towardsdatascience.com/making-publication-quality-figures-in-python-part-ii-line-plot-legends-colors-4430a5891706

```python
# load packages
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

# modify global setting
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
mpl.rcParams['font.family'] = 'Arial'

# prepare some data
np.random.seed(42)
x = np.arange(1,11,1)
y1 = np.linspace(300,400,10)
y2 = np.random.randint(low=300,high=400,size=10)
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0.1,0.1,0.5,0.8])#左下起点，width，height
ax2 = fig.add_axes([0.15,0.6,0.2,0.2])
ax.plot(x,y1,
       marker = '*',
       markersize = 8,
       markerfacecolor = 'red',
       markeredgecolor = 'black',
       markeredgewidth = '2',
       linestyle ='--',
       linewidth = 3)
```

```python
#fmt='ro-' which is equivalent to marker='o',markerfacecolor='red',linestype='-'

fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0.1,0.1,0.5,0.8])
ax.plot(x,y1,
       'ro-',
       #marker = '*',
       markersize = 8,
       #markerfacecolor = 'red',
       markeredgecolor = 'black',
       markeredgewidth = '2',
       #linestyle ='--',
       linewidth = 3)
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0.1,0.1,0.5,0.8])
ax.plot(x,y1,marker='o',markersize=8,markerfacecolor='red',markeredgecolor='black',markeredgewidth=2,linestyle='--',linewidth=3)
ax.plot(x,y2,marker='o')
```

```python
# the higher the zorder is, the later the line would be drawn
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0.1,0.1,0.5,0.8])
ax.plot(x,y1,marker='o',markersize=8,markerfacecolor='red',markeredgecolor='black',markeredgewidth=2,linestyle='--',linewidth=3,zorder=3)
ax.plot(x,y2,marker='o')

ax.legend()
```

```python
# the higher the zorder is, the later the line would be drawn
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0.1,0.1,0.5,0.8])
p1 = ax.plot(x,y1,marker='o',markersize=8,markerfacecolor='red',markeredgecolor='black',markeredgewidth=2,linestyle='--',linewidth=3,zorder=3)
p2 = ax.plot(x,y2,marker='o')

ax.legend(handles = [p1[0],p2[0]],labels = ['blueline','orangeline'])
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0.1,0.1,0.5,0.8])
h1 = ax.plot([],[],marker='o',markersize=8,markerfacecolor='red',markeredgecolor='black',markeredgewidth=2,linestyle='--',linewidth=3)
h2 = ax.plot([],[],marker='o')
ax.legend(handles=[h1[0],h2[0]],labels=['blue line','orange line'])
```

```python
# the higher the zorder is, the later the line would be drawn
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0.1,0.1,0.5,0.8])
p1 = ax.plot(x,y1,marker='o',markersize=8,markerfacecolor='red',markeredgecolor='black',markeredgewidth=2,linestyle='--',linewidth=3,zorder=3)
p2 = ax.plot(x,y2,marker='o')

ax.legend(handles = [p1[0],p2[0]],labels = ['blueline','orangeline'],
         loc = 'upper right',
         bbox_to_anchor = (1.3,1),
         title = 'legend',
         frameon = False)

```

## 添加子图的几种方式
https://blog.csdn.net/fengdu78/article/details/107888652

### 方式一：通过plt的subplot

```python
x=np.arange(1,100)
plt.subplot(221)
plt.plot(x,x*x)
plt.subplot(222)
plt.scatter(np.arange(0,10), np.random.rand(10))
# 画第3个图：饼图
plt.subplot(223)
plt.pie(x=[15,30,45,10],labels=list('ABCD'),autopct='%.0f',explode=[0,0.05,0,0])
# 画第4个图：条形图
plt.subplot(224)
plt.bar([20,10,30,25,15],[25,15,35,30,20],color='b')
```

### 方式二：通过figure的add_subplot

```python
fig = plt.figure(figsize=(10,6))
ax1 = fig.add_subplot(2,2,1)
ax1.plot(x,x*x)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
```

### 方式三：通过plt的subplots

```python
#subplots返回的值的类型为元组，其中包含两个元素：第一个为一个画布，第二个是子图

fig,axes=plt.subplots(2,2)
# 画第1个图：折线图
x=np.arange(1,100)
axes[0][0].plot(x,x*x)
# 画第2个图：散点图
```

```python
import pandas as pd
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import Normalize
from scipy.stats import gaussian_kde
from matplotlib import rcParams
config = {"font.family": 'Times New Roman',
          "font.size": 16, "mathtext.fontset": 'stix'}
rcParams.update(config)
# 读取数据

#filename = r'F:/Rpython/lp37/testdata.xlsx'
filename = r'D:/Project/pan_qaidam_yardang/0_grid/csv/grid_exclude_non_yardang.csv'
df2 = pd.read_csv(filename)  # 读取文件
x = df2['NUMPOINTS'].values.ravel()
y = df2['TRI90'].values.ravel()
N = len(df2['NUMPOINTS'])
# 绘制拟合线
x2 = np.linspace(-10, 30)
y2 = x2


def f_1(x, A, B):
  return A*x + B


A1, B1 = optimize.curve_fit(f_1, x, y)[0]
y3 = A1*x + B1
# Calculate the point density
xy = np.vstack([x, y])
z = gaussian_kde(xy)(xy)
norm = Normalize(vmin=np.min(z), vmax=np.max(z))
# 开始绘图
fig, ax = plt.subplots(figsize=(12, 9), dpi=600)
scatter = ax.scatter(x, y, marker='o', c=z*100, 
                     #edgecolors='',
                     s=15, label='LST', cmap='Spectral_r')
cbar = plt.colorbar(scatter, shrink=1, orientation='vertical',
                    extend='both', pad=0.015, aspect=30, label='frequency')
cbar.ax.locator_params(nbins=8)
# 0,0.005,0.010,0.015,0.020,0.025,0.030,0.035
cbar.ax.set_yticklabels([0.005, 0.010, 0.015, 0.020, 0.025, 0.030, 0.035])
ax.plot(x2, y2, color='k', linewidth=1.5, linestyle='--')
ax.plot(x, y3, color='r', linewidth=2, linestyle='-')
fontdict1 = {"size": 16, "color": "k", 'family': 'Times New Roman'}
ax.set_xlabel("PRE", fontdict=fontdict1)
ax.set_ylabel("OBS", fontdict=fontdict1)
# ax.grid(True)
ax.set_xlim((0, 25))
ax.set_ylim((0, 25))
ax.set_xticks(np.arange(0, 25.1, step=5))
ax.set_yticks(np.arange(0, 25.1, step=5))
plt.savefig('plot72.png', dpi=800,
            bbox_inches='tight', pad_inches=0)
plt.show()
```
