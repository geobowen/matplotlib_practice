## Making publication-quality figures in Python (Part I): Fig and Axes
https://towardsdatascience.com/making-publication-quality-figures-in-python-part-i-fig-and-axes-d86c3903ad9b

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter, MaxNLocator
```

```python
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
mpl.rcParams['font.family'] = 'Arial'
```

```python
fig = plt.figure(figsize=(10,6))
ax = plt.axes((0.1, 0.1, 0.5, 0.8))

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(True)

ax.tick_params(axis='x', 
               which = 'major', 
               direction = 'out',
               length = 10, 
               width = 5, 
               color = 'red', 
               pad = 15,
               labelsize = 15, 
               labelcolor = 'green', 
               labelrotation = 15)

ax.set_xticks([0.2,0.8,1])
ax.set_xticklabels(['pos1','pos2','pos3'])

#设置主刻度（major tick）间隔
ax.yaxis.set_major_locator(MultipleLocator(0.1))
ax.yaxis.set_minor_locator(MultipleLocator(0.05))
```

## Making publication-quality figures in python (Part II): Line plot, Legends, Colors
https://towardsdatascience.com/making-publication-quality-figures-in-python-part-ii-line-plot-legends-colors-4430a5891706

```python
# load packages
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

# modify global setting
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
mpl.rcParams['font.family'] = 'Arial'

# prepare some data
np.random.seed(42)
x = np.arange(1,11,1)
y1 = np.linspace(300,400,10)
y2 = np.random.randint(low=300,high=400,size=10)
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0.1,0.1,0.5,0.8])#左下起点，width，height
ax2 = fig.add_axes([0.15,0.6,0.2,0.2])
ax.plot(x,y1,
       marker = '*',
       markersize = 8,
       markerfacecolor = 'red',
       markeredgecolor = 'black',
       markeredgewidth = '2',
       linestyle ='--',
       linewidth = 3)
```

```python
#fmt='ro-' which is equivalent to marker='o',markerfacecolor='red',linestype='-'

fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0.1,0.1,0.5,0.8])
ax.plot(x,y1,
       'ro-',
       #marker = '*',
       markersize = 8,
       #markerfacecolor = 'red',
       markeredgecolor = 'black',
       markeredgewidth = '2',
       #linestyle ='--',
       linewidth = 3)
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0.1,0.1,0.5,0.8])
ax.plot(x,y1,marker='o',markersize=8,markerfacecolor='red',markeredgecolor='black',markeredgewidth=2,linestyle='--',linewidth=3)
ax.plot(x,y2,marker='o')
```

```python
# the higher the zorder is, the later the line would be drawn
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0.1,0.1,0.5,0.8])
ax.plot(x,y1,marker='o',markersize=8,markerfacecolor='red',markeredgecolor='black',markeredgewidth=2,linestyle='--',linewidth=3,zorder=3)
ax.plot(x,y2,marker='o')

ax.legend()
```

```python
# the higher the zorder is, the later the line would be drawn
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0.1,0.1,0.5,0.8])
p1 = ax.plot(x,y1,marker='o',markersize=8,markerfacecolor='red',markeredgecolor='black',markeredgewidth=2,linestyle='--',linewidth=3,zorder=3)
p2 = ax.plot(x,y2,marker='o')

ax.legend(handles = [p1[0],p2[0]],labels = ['blueline','orangeline'])
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0.1,0.1,0.5,0.8])
h1 = ax.plot([],[],marker='o',markersize=8,markerfacecolor='red',markeredgecolor='black',markeredgewidth=2,linestyle='--',linewidth=3)
h2 = ax.plot([],[],marker='o')
ax.legend(handles=[h1[0],h2[0]],labels=['blue line','orange line'])
```

```python
# the higher the zorder is, the later the line would be drawn
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0.1,0.1,0.5,0.8])
p1 = ax.plot(x,y1,marker='o',markersize=8,markerfacecolor='red',markeredgecolor='black',markeredgewidth=2,linestyle='--',linewidth=3,zorder=3)
p2 = ax.plot(x,y2,marker='o')

ax.legend(handles = [p1[0],p2[0]],labels = ['blueline','orangeline'],
         loc = 'upper right',
         bbox_to_anchor = (1.3,1),
         title = 'legend',
         frameon = False)

```

## 添加子图的几种方式
https://blog.csdn.net/fengdu78/article/details/107888652

### 方式一：通过plt的subplot

```python
x=np.arange(1,100)
plt.subplot(221)
plt.plot(x,x*x)
plt.subplot(222)
plt.scatter(np.arange(0,10), np.random.rand(10))
# 画第3个图：饼图
plt.subplot(223)
plt.pie(x=[15,30,45,10],labels=list('ABCD'),autopct='%.0f',explode=[0,0.05,0,0])
# 画第4个图：条形图
plt.subplot(224)
plt.bar([20,10,30,25,15],[25,15,35,30,20],color='b')
```

### 方式二：通过figure的add_subplot

```python
fig = plt.figure(figsize=(10,6))
ax1 = fig.add_subplot(2,2,1)
ax1.plot(x,x*x)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
```

### 方式三：通过plt的subplots

```python
#subplots返回的值的类型为元组，其中包含两个元素：第一个为一个画布，第二个是子图

fig,axes=plt.subplots(2,2)
# 画第1个图：折线图
x=np.arange(1,100)
axes[0][0].plot(x,x*x)
# 画第2个图：散点图
```

```python
fig = plt.figure()
gs = mpl.gridspec.GridSpec(nrows=3,
                           ncols=2,
                           width_ratios=(0.4,0.6),
                           height_ratios=(0.1,0.1,0.8),
                           hspace=0.3)
ax1 = fig.add_subplot(gs[0,1])
ax2 = fig.add_subplot(gs[1,1])
ax3 = fig.add_subplot(gs[2,1])
ax4 = fig.add_subplot(gs[2,0])
ax5 = fig.add_subplot(gs[0,0])
```

```python
fig, ax = plt.subplots(40, 10, figsize=(5, 20))
fig.subplots_adjust(hspace=0, wspace=0)

# Get some face data from scikit-learn
from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces().images

for i in range(40):
    for j in range(10):
        ax[i, j].xaxis.set_major_locator(plt.NullLocator())
        ax[i, j].yaxis.set_major_locator(plt.NullLocator())
        ax[i, j].imshow(faces[10 * i + j], 
                        cmap="bone"
                       )
```

```python
from matplotlib.backends.backend_agg import FigureCanvasAgg
from matplotlib.figure import Figure
from matplotlib.axes import Axes
import numpy as np
from numpy import math
fig =Figure( 
 dpi=100,
 facecolor=(239/256,239/256,239/256),
 edgecolor=(82/256,101/256,155/256),
 linewidth=6.0,
 frameon=True,
 )
canvas = FigureCanvasAgg(fig)
#创建一个2*2的网格，并在第4个格子中创建一个axes
fig.add_subplot(224)
s, (width, height) = canvas.print_to_buffer() 
from PIL import Image 
im = Image.frombytes("RGBA", (width, height), s)
im.show() 
```
